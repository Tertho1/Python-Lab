```python
df_pivot = df.pivot_table(values='CO2EMISSIONS', index='CYLINDERS', columns='FUELTYPE', aggfunc='mean')
print("\nPivot Table of CO2 Emissions:")
print(df_pivot)
```

```python
df['CO2EMISSIONS'] = df['CO2EMISSIONS'].apply(lambda x: x / 1000)
print("\nData with CO2 Emissions converted to kg/km:")
print(df[['CO2EMISSIONS', 'CO2EMISSIONS']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('/content/FuelConsumptionCo2.csv')
plt.figure(figsize=(8, 6))
plt.scatter(df['ENGINESIZE'], df['FUELCONSUMPTION_COMB'], color='blue', alpha=0.7)

plt.title('Fuel Consumption vs Engine Size')
plt.xlabel('Engine Size (Liters)')
plt.ylabel('Fuel Consumption (L/100km)')

plt.grid(True)
plt.show()
```
